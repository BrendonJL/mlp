```python
# EXACT copy from working random_agent.py
import matplotlib.pyplot as plt
import gym_super_mario_bros

# Create the Environment (same as script)
env = gym_super_mario_bros.make("SuperMarioBros-v3", apply_api_compatibility=True)

# Reset (no tuple! Just like script)
observation = env.reset()

# Collect frames
num_frames = 6
frames = []

for i in range(num_frames):
    action = env.action_space.sample()
    # 5 values - exactly like script!
    observation, reward, terminated, truncated, info = env.step(action)
    frames.append(observation)

    done = terminated or truncated
    if done:
        observation = env.reset()

print(f"SUCCESS! Collected {len(frames)} frames")

# Display grid
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for idx, ax in enumerate(axes.flat):
    ax.imshow(frames[idx])
    ax.set_title(f"Frame {idx + 1}")
    ax.axis('off')
plt.show()
```
